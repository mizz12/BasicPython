## 1問目

```python
hist_data = pd.read_csv('/Users/miz/Desktop/PG/BasicPython/data/hist_data.csv')
```

```python
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")

plt.hist(hist_data["aデータ"], bins = 30, alpha=0.5, label = "aデータ", color = "orange")
plt.hist(hist_data["bデータ"], bins = 30, alpha=0.3, label = "bデータ", color = "c")
plt.legend(loc="upper right")
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
scatter_data = pd.read_csv("/Users/miz/Desktop/PG/BasicPython/data/scatter_data.csv")
df0 = scatter_data[scatter_data["class"] == 0]
df1 = scatter_data[scatter_data["class"] == 1]
df2 = scatter_data[scatter_data["class"] == 2]
```

```python
plt.title("クラス0~2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")

plt.scatter(df0["length (cm)"],df0["width (cm)"], label = "class0", color = "r")
plt.scatter(df1["length (cm)"],df1["width (cm)"], label = "class1", color = "b")
plt.scatter(df2["length (cm)"],df2["width (cm)"], label = "class2", color = "g")
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

```python
line_data = pd.read_csv("/Users/miz/Desktop/PG/BasicPython/data/line_data.csv")
```

```python
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("件数")

x = line_data["期間"]
y = line_data["件数"]

plt.plot(x,y,marker = "o")

plt.xticks(rotation=20)
plt.grid()
plt.show()

plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")

x = line_data["期間"]
y = line_data["費用"]

plt.bar(x,y)

plt.xticks(rotation=20)
plt.grid()
plt.show()
```
